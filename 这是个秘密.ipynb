```python
import matplotlib.pyplot as plt
from matplotlib import animation
import numpy as np
import math
```

```python
t = np.linspace(0,math.pi,1000)
x = np.sin(t)
y = np.cos(t) + np.power(x,2/4)
plt.scatter(x,y,c=y,cmap = plt.cm.Reds, edgecolor = 'none',s = 100)
plt.scatter(-x,y,c=y,cmap = plt.cm.Reds, edgecolor = 'none',s = 100)
plt.fill(x,y,'r',alpha = 1)
plt.fill(-x,y,'r',alpha = 1)
plt.axis([-2,2,-1.5,2])
plt.title('I love you, Baby', fontsize = 30)
plt.show()
```
